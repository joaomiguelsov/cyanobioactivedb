```python
import pandas as pd 
import numpy as np
```

```python
activities=pd.read_csv("Compound_Chembl_Activity.csv",sep=";")
activities_IC50=activities[activities["type"]=="IC50"]
activities_IC50
```

```python
activities_IC50=activities_IC50.replace("None",np.nan)
activities_IC50["value"]=activities_IC50["value"].replace("None",np.nan)
```

```python
value_number=[]

for value in activities_IC50["value"]:
    value_number.append(float(value))
activities_IC50['value']=value_number
```

```python
activities_IC50['value'] = np.where(activities_IC50['units'] == 'uM', activities_IC50['value'] * 1000, activities_IC50['value'])
```

```python
activities_IC50['value'] = np.where(activities_IC50['units'] == 'um', activities_IC50['value'] * 1000, activities_IC50['value'])
```

```python
activities_IC50['value'] = np.where(activities_IC50['units'] == 'pM', activities_IC50['value'] * 0.001, activities_IC50['value'])
```

```python
#activities_IC50["value"]=activities["value"].astype(str)

value_string=[]

for value in activities_IC50["value"]:
    value_string.append(str(value))
    
activities_IC50['value']=value_string

activities_IC50
```

```python
activities_IC50["units"]=activities_IC50["units"].replace("uM","nM")
activities_IC50["units"]=activities_IC50["units"].replace("um","nM")
activities_IC50["units"]=activities_IC50["units"].replace("pM","nM")
```

```python
def create_activity_column(name_of_column,type_of_activity):
    activities_IC50[str(name_of_column)]=activities_IC50.query('type==@type_of_activity')["value"]+activities_IC50["units"]
    activities_IC50[str(name_of_column)]=activities_IC50[str(name_of_column)].fillna("None")
    activities_IC50[str(name_of_column)]= activities_IC50[str(name_of_column)].astype("object")
```

```python
create_activity_column("IC50","IC50")
```

```python
activities_IC50=activities_IC50.sort_values("target_chembl_id")
activities_IC50=activities_IC50.dropna(subset=["value"])
activities_IC50=activities_IC50.reset_index()
activities_IC50
```

```python
activities_IC50=activities_IC50.loc[activities_IC50["units"]=="nM"]
activities_IC50
```

```python
chembl_column=activities_IC50["molecule_chembl_id"]
```

```python

activities_IC50 = activities_IC50.groupby("molecule_chembl_id", as_index=False)[['activity_comment', 'activity_id', 'activity_properties',
       'assay_chembl_id', 'assay_description', 'assay_type',
       'assay_variant_accession', 'assay_variant_mutation', 'bao_endpoint',
       'bao_format', 'bao_label', 'canonical_smiles', 'data_validity_comment',
       'data_validity_description', 'document_chembl_id', 'document_journal',
       'document_year', 'ligand_efficiency', 'molecule_pref_name',
       'parent_molecule_chembl_id', 'pchembl_value', 'potential_duplicate',
       'qudt_units', 'record_id', 'relation', 'src_id', 'standard_flag',
       'standard_relation', 'standard_text_value', 'standard_type',
       'standard_units', 'standard_upper_value', 'standard_value',
       'target_chembl_id', 'target_organism', 'target_pref_name',
       'target_tax_id', 'text_value', 'toid', 'type', 'units', 'uo_units',
       'upper_value', 'value','IC50']].transform(lambda x: '||'.join(x))
```

```python
activities_IC50
```

```python
activities_IC50["chembl_id"]=chembl_column
activities_IC50["target_chembl_IDs"]=activities_IC50["target_chembl_id"]
activities_IC50
```

```python
activities_IC50=activities_IC50.drop_duplicates(subset="chembl_id")
activities_IC50=activities_IC50.drop(activities_IC50.iloc[:,0:15],axis=1)
activities_IC50.reset_index()
activities_IC50
```
