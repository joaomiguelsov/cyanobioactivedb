```python
import drugtax
import pandas as pd
import numpy as np
```

```python
class drugtax_of_database:
    def __init__(self,nome_do_ficheiro,indice_do_ficheiro,nome_do_output):
        self.nome_do_ficheiro=nome_do_ficheiro
        self.indice_do_ficheiro =indice_do_ficheiro
        self.nome_do_output = nome_do_output

    def run_drugtax(self):
        file=pd.read_csv(self.nome_do_ficheiro,sep=";",encoding="UTF-8")
        #smiles=file[self.indice_do_ficheiro]
        smiles=[]
        for smile in file[self.indice_do_ficheiro]:
            smiles.append(smile)
        count=0
        features_list=[]
        for x in smiles:
            while count<1:
                count+=1
                molecule1=drugtax.DrugTax(x)
                features=molecule1.features
                for feature in features:
                    features_list.append(str(feature))

        output=open(self.nome_do_output,"w")
        output.write("SMILE;Kingdom;Superclasses;")
        for feature in features_list:
            output.write(feature +";")
        output.write("\n")
        for smile in smiles:
            molecule=drugtax.DrugTax(str(smile))
            results = str(molecule)
            print(molecule)
            features = molecule.features
            results2 = results.replace("\n", ";")
            results3 = results2.replace("SMILE:", "")
            results4 = results3.replace("Kingdom:", "")
            results5 = results4.replace("Superclasses:", "")
            output.write(results5)
            #print(results5)
            for feature in features:
                output.write(str(features[feature]) + ";")
            output.write("\n")
        output.close()
        return
```

```python
final_version=drugtax_of_database("onlyIC50.csv","Smiles_Isomeric","Results_final_Drugtax.txt")
final_version.run_drugtax()
```

```python
cyano_Drugtax["IC50"]=cyano_Drugtax["IC50"].astype(str)
cyano_Drugtax["target_chembl_IDs"]=cyano_Drugtax["target_chembl_IDs"].astype(str)
```

```python
cyano_Drugtax=cyano_Drugtax.iloc[:,:-1]
cyano_Drugtax_exploded=(cyano_Drugtax.assign(Individual_targets=cyano_Drugtax["target_chembl_IDs"].str.split("\|\|")).explode("Individual_targets").reset_index(drop=True))
cyano_Drugtax_exploded=cyano_Drugtax_exploded.drop("Individual_targets",axis=1)
```

```python
Exploded_rows = (cyano_Drugtax.apply(lambda row: pd.Series(list(zip(row['IC50'].split('||'), row['target_chembl_IDs'].split('||')))), axis=1)
                  .stack()
                  .apply(pd.Series)
                  .reset_index(drop=True)
                  .rename(columns={0: 'IC50_individual_values', 1: 'Individual_targets'}))


cyano2=pd.concat([cyano_Drugtax_exploded,Exploded_rows],axis=1)
```

```python
cyano2.loc[cyano2['IC50_individual_values']=="nan",'IC50_individual_values']=np.nan
cyano2.loc[cyano2['Individual_targets']=="nan",'Individual_targets']=np.nan
```

```python
T=cyano2.groupby("Individual_targets")["Individual_targets"].count()
T
```

```python
cyano2=cyano2[cyano2.Individual_targets.isin(T.index[T.ge(25)])]
cyano2=cyano2.dropna(subset="Individual_targets")
```
