```python
import pandas as pd
import numpy as np
```

```python
cyano=pd.read_csv("onlyIC50__Mordred_results.csv",sep=";")
cyano
```

```python
cyano["IC50"]=cyano["IC50"].astype(str)
cyano["target_chembl_IDs"]=cyano["target_chembl_IDs"].astype(str)
```

```python
cyano_exploded=(cyano.assign(Individual_targets=cyano["target_chembl_IDs"].str.split("\|\|")).explode("Individual_targets").reset_index(drop=True))
```

```python
cyano_exploded=cyano_exploded.drop("Individual_targets",axis=1)
```

```python
Exploded_rows = (cyano.apply(lambda row: pd.Series(list(zip(row['IC50'].split('||'), row['target_chembl_IDs'].split('||')))), axis=1)
                  .stack()
                  .apply(pd.Series)
                  .reset_index(drop=True)
                  .rename(columns={0: 'IC50_individual_values', 1: 'Individual_targets'}))


cyano2=pd.concat([cyano_exploded,Exploded_rows],axis=1)
```

```python
cyano2.loc[cyano2['IC50_individual_values']=="nan",'IC50_individual_values']=np.nan
cyano2.loc[cyano2['Individual_targets']=="nan",'Individual_targets']=np.nan
```

```python
pd.set_option("display.max_rows",100)
#cyano2["IC50_individual_values"]=cyano2["IC50_individual_values"].str.replace("uM","")
cyano2
```

```python
cyano2["IC50_individual_values"].unique()
```

```python
T=cyano2.groupby("Individual_targets")["Individual_targets"].count()
T
```

```python
cyano2=cyano2[cyano2.Individual_targets.isin(T.index[T.ge(25)])]

cyano2=cyano2.dropna(subset="Individual_targets")
cyano2
```

```python
T=cyano2.groupby("Individual_targets")["Individual_targets"].count()
T
```

```python
cyano2=cyano2.reset_index()
```

```python
cyano2
```

```python
cyano_no_results=cyano2.iloc[:,:20]
cyano_no_results
```

```python
cyano_results=cyano2.iloc[:,20:-2]
#cyano_results=cyano_results.drop(["Lipinski","GhoseFilter"],axis=1)
cyano_results=cyano_results.replace(to_replace="\D",value=np.nan,regex=True)
cyano_results=cyano_results.loc[:,cyano_results.isnull().mean() < 0.2]
cyano_results
```

```python
targets=cyano2.loc[:,"Individual_targets"]
values=cyano2.loc[:,"IC50_individual_values"]
```

```python
cyano2=pd.concat([cyano_no_results,cyano_results],axis=1)

cyano2=pd.concat([cyano2,values],axis=1)
cyano2=pd.concat([cyano2,targets],axis=1)

cyano2=cyano2.drop("index",axis=1)
cyano2
```

```python
cyano2.dropna(subset="Individual_targets")
```

```python
not_acepted_values=["None","nanug ml-1"]

cyano2= cyano2[~cyano2["IC50_individual_values"].isin(not_acepted_values)]
cyano2
```

```python
cyano2.to_csv("Mordred_ML_No_resample_v2.csv", sep=";", index=False)
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
final_db=cyano2.drop("Individual_targets",axis=1)
```

```python
ros=RandomOverSampler()
x_resample, y_resample=ros.fit_resample(final_db,targets)
```

```python
x_resample["Individual_targets"]=y_resample
x_resample
```

```python
x_resample.to_csv("Mordred_Resample_v2.csv",sep=";",index=False)
```
