```python
import pandas as pd
import numpy as np
```

```python
padel_desc=pd.read_csv("CyanoBioactiveDB_Complete_PadelDescriptors.csv",sep=";")
padel_desc
```

```python
padel_desc=padel_desc.loc[:,padel_desc.isnull().mean() < 0.2]
padel_desc
```

```python
cyano=pd.read_csv("onlyIC50.csv",sep=";")
cyano
```

```python
cyano_padel=pd.merge(cyano,padel_desc, how="inner",left_on="Compound",right_on="Name")
cyano_padel
```

```python
cyano_padel.to_csv("Padel_IC50.csv",sep=";",index=False)
```

```python
cyano_padel["IC50"]=cyano_padel["IC50"].astype(str)
cyano_padel["target_chembl_IDs"]=cyano_padel["target_chembl_IDs"].astype(str)
```

```python
cyano_padel_exploded=(cyano_padel.assign(Individual_targets=cyano_padel["target_chembl_IDs"].str.split("\|\|")).explode("Individual_targets").reset_index(drop=True))
#cyano_padel2=cyano_padel2.drop_duplicates()
cyano_padel_exploded=cyano_padel_exploded.drop("Individual_targets", axis=1)
```

```python
Exploded_rows = (cyano_padel.apply(lambda row: pd.Series(list(zip(row['IC50'].split('||'), row['target_chembl_IDs'].split('||')))), axis=1)
                  .stack()
                  .apply(pd.Series)
                  .reset_index(drop=True)
                  .rename(columns={0: 'IC50_individual_values', 1: 'Individual_targets'}))


cyano2=pd.concat([cyano_padel_exploded,Exploded_rows],axis=1)
```

```python
cyano2
```

```python
cyano2.loc[cyano2['IC50_individual_values']=="nan",'IC50_individual_values']=np.nan
cyano2.loc[cyano2['Individual_targets']=="nan",'Individual_targets']=np.nan
```

```python
cyano2
```

```python
T=cyano2.groupby("Individual_targets")["Individual_targets"].count()
T
```

```python
cyano2=cyano2[cyano2.Individual_targets.isin(T.index[T.ge(25)])]
cyano2
```

```python
targets=cyano2.loc[:,"Individual_targets"]
```

```python
cyano2.to_csv("Padel_IC50_ML_v2_noresample.csv",sep=";", index=False)
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
final_db=cyano2.drop("Individual_targets",axis=1)
```

```python
ros=RandomOverSampler()
x_resample, y_resample=ros.fit_resample(final_db,targets)
```

```python
x_resample["Individual_targets"]=y_resample
x_resample
```

```python
x_resample.to_csv("Padel_IC50_ML_v2_resample.csv",sep=";",index=False)
```
