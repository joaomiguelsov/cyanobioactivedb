```python
from rdkit import Chem
from mordred import Calculator, descriptors
import pandas as pd
import numpy
```

```python
# create descriptor calculator with all descriptors
calc = Calculator(descriptors, ignore_3D=True)
```

```python
cyanoter=pd.read_csv("onlyIC50.csv",sep=";")

cyanoter
```

```python
cyanoter2=cyanoter.dropna(subset=["Smiles_Isomeric"])
cyanoter2=cyanoter2.reset_index(drop=True)
cyanoter2
```

```python
lista = cyanoter2.Smiles_Isomeric.astype(str).to_list()
lista
```

```python
len(lista)
```

```python
 mols = [Chem.MolFromSmiles(smi) for smi in cyanoter2.Smiles_Isomeric]
```

```python
df = calc.pandas(mols)
df
```

```python
#pd.set_option("display.max_columns",3000)
df
```

```python
cyanoter2
```

```python
pd.set_option("display.max_columns",300)
```

```python
cyanoter2_descriptors=pd.concat([cyanoter2,df],axis=1,join="inner")
cyanoter2_descriptors
```
