```python
import pandas as pd
import numpy as np


pd.set_option("display.max_columns",200)          #allows us to see up to 200 columns without skypping them.
#pd.set_option('display.max_rows', 10000)         #allows us to see up to 10000 rows without skypping them.
```

```python
cyano = pd.read_csv("CyanometDB_Curated.csv",sep=";")        #Reads the csv file for the cyanomet database
cyano
```

```python
cyano["CyanoMetDB"]="True"                                  #Creates a column in the database with all rows being true
```

```python
cyano
```

```python
pubchem=pd.read_csv("PUBCHEM_Database_BIOASSAYS_28102022.csv",sep=";")             #Reads the csv file for the pubchem database

pubchem
```

```python
#Renames the columns as stated


pubchem = pubchem.rename(columns={'PUBCHEM_OPENEYE_CAN_SMILES': 'SMILES (canonical)', 
                                  'PUBCHEM_OPENEYE_ISO_SMILES': 'SMILES (isomeric)',"PUBCHEM_IUPAC_INCHI":"InChI","PUBCHEM_IUPAC_INCHIKEY" : "InChlKey"})


#creates a column in the database with all rows being true


pubchem["PubChemDB"]="True"
pubchem
```

```python
Atlas=pd.read_csv("Atlas_final.csv",sep=";")      #Reads the csv file for the Atlas database


Atlas["NPAtlasDB"]="True"                         #creates a column in the database with all rows being true


Atlas
```

## Bellow we start the merging of the databases by the InchIkey column on both databases the "outer" method allows us to mantain all the rows\information in the databases.

```python
cyano_pubchem=pd.merge(cyano,pubchem,left_on=["InChlKey"],right_on=["InChlKey"], how="outer")

cyano_pubchem
```

```python
#Creates a new column called InchI_all by joiningvalues of the two InchI columns of the database(dont confuse Inchi with InchiKey)
cyano_pubchem["InChI_all"] = cyano_pubchem['InChI_x'].fillna(cyano_pubchem["InChI_y"])
```

```python
#Renames the stated column to InChIKey
Atlas = Atlas.rename(columns={"compound_inchikey" : "InChlKey"})
```

```python
# same method as stated before
cyano_pubchem_Atlas=pd.merge(cyano_pubchem,Atlas,left_on=["InChlKey"],right_on=["InChlKey"],how="outer")


cyano_pubchem_Atlas
```

```python
# Same as before
cyano_pubchem_Atlas["InChI_all"] = cyano_pubchem_Atlas['InChI_all'].fillna(cyano_pubchem_Atlas["InChI_y"])
```

```python
cyano_pubchem_Atlas
```

```python
#Joins the canonical smiles from the cyanometdb and pubchemdb into only one column mantaining the cyanometDB Smiles where it is available an filling the blank spaces with smiles from pubchem.
cyano_pubchem_Atlas["All_SMILES (canonical)"]=cyano_pubchem_Atlas['SMILES (canonical)_x'].fillna(cyano_pubchem_Atlas["SMILES (canonical)_y"])


#Joins the canonical Smiles from before with the ones from the NPatlasDB only filling missing information in the rows when possible.
cyano_pubchem_Atlas["All_SMILES (canonical)"]=cyano_pubchem_Atlas['All_SMILES (canonical)'].fillna(cyano_pubchem_Atlas["compound_smiles"])
cyano_pubchem_Atlas
```

```python
#Creates a copy of the Fused databases fillinfg the rows that identify in which database they originated from with the remaining information 


all_merges=cyano_pubchem_Atlas.copy()
all_merges["CyanoMetDB"]=all_merges["CyanoMetDB"].fillna("False")
all_merges["PubChemDB"]=all_merges["PubChemDB"].fillna("False")
all_merges["NPAtlasDB"]=all_merges["NPAtlasDB"].fillna("False")
all_merges
```

## Here we start defining the final columns that will be in our final database

```python
all_merges["Compound_name"]=all_merges['Compound name'].fillna(all_merges['compound_names'])
all_merges['Classes_of_compounds']=all_merges['Class of compound']
all_merges['Genus_of_origin']=all_merges["Genus"].fillna(all_merges["genus"])
all_merges['Smiles_Canonical']=all_merges['All_SMILES (canonical)']
all_merges['Smiles_Isomeric']=all_merges['SMILES (isomeric)_x'].fillna(all_merges['SMILES (isomeric)_y'])
all_merges['InChlKey_final']=all_merges['InChlKey']
all_merges['InChl']=all_merges["InChI_all"]
all_merges["DOI"]=all_merges['DOI_No1'].fillna(all_merges['original_reference_doi'])
all_merges["URL"]=all_merges['npatlas_url']
all_merges["CyanoMet_DB"]=all_merges["CyanoMetDB"]
all_merges["PubChem_DB"]=all_merges["PubChemDB"]
all_merges["NPAtlas_DB"]=all_merges["NPAtlasDB"]


cyanobioativeDB=all_merges.drop(all_merges.iloc[:,0:93],axis=1)
```

```python
cyanobioativeDB
```

```python
#Compounds 69 to 76 and 339 and 2090 of the original Cyanomet database cant find inchikey so removed them
cyanobioativeDB=cyanobioativeDB.dropna(axis=0,subset=["InChlKey_final"])
cyanobioativeDB
```

```python
#Droping the duplicates and reseting the index so the count matchs up
cyanobioativeDB=cyanobioativeDB.drop_duplicates(subset=["InChlKey_final"])
cyanobioativeDB=cyanobioativeDB.reset_index(drop=True)
cyanobioativeDB
```

## Here we compile our own index for the compound and replace irregular entries found in the genus column 

```python
lcomp=[]
for x in range(1,len(cyanobioativeDB)+1):
    lcomp.append("Cyano_"+str(x).zfill(4))
```

```python
cyanobioativeDB["Compound"]=lcomp
cyanobioativeDB.insert(0,"Compound",cyanobioativeDB.pop("Compound"))
```

```python
print(cyanobioativeDB.Genus_of_origin.unique())
```

```python
cyanobioativeDB["Genus_of_origin"]=cyanobioativeDB["Genus_of_origin"].replace("cf. Oscillatoria and Hormoscilla ","Oscillatoria/Hormoscilla")
cyanobioativeDB["Genus_of_origin"]=cyanobioativeDB["Genus_of_origin"].replace(" sp.","")
cyanobioativeDB["Genus_of_origin"]=cyanobioativeDB["Genus_of_origin"].replace("Unknown",np.nan)
cyanobioativeDB["Genus_of_origin"]=cyanobioativeDB["Genus_of_origin"].replace("Aetokthonos hydrillicola","Aetokthonos")
cyanobioativeDB["Genus_of_origin"]=cyanobioativeDB["Genus_of_origin"].replace("field sample: Microcystis bloom","Microcystis")
cyanobioativeDB["Genus_of_origin"]=cyanobioativeDB["Genus_of_origin"].replace("Desmonostoc sp.","Desmonostoc")
```

```python
cyanobioativeDB["Smiles_Isomeric"]=cyanobioativeDB["Smiles_Isomeric"].replace("none",np.nan)
```

```python
cyanobioativeDB
```

```python
cyanobioativeDB.to_csv("Cyanoter_V1.csv",sep=";",index=False)              #Saves database to a csv file
```
