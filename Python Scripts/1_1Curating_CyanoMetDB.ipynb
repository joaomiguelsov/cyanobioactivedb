## Initially we need to curate information from the collected databases in this case CyanoMetDB

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns",200) #allows us to see up to 200 columns without skypping them.
```

```python
cyanodb=pd.read_csv("CyanoMetDB_v02_2023.csv",sep=";",encoding='latin-1') #reads the csv file in thos case encoding wasnt working with utf-8 so used the latin-1 encoding
cyanodb
```

```python
#drops the respective named columns from the database
#"reason why not included in analysis/count"


cyanodb=cyanodb.drop(["Reference_Text_No1 Title; Journal; Vol,; Issue; pages; year; type; DOI; author1; authors2; etc.","Bibtex No1","Ref_ID_No2","DOI_No2","Url_No2",
                      "Reference_Text_No2 No Title; Journal; Vol,; Issue; pages; year; type; DOI; author1; authors2; etc.", "PubMed ID No2","Bibtex No2","Ref_ID_No3","DOI_No3",
                      "Url_No3","Reference_Text_No3 Title; Journal; Vol,; Issue; pages; year; type; DOI; author1; authors2; etc.", "PubMed ID No3","Bibtex No3","Build block string",
                      "Strain","Species","Field sample","PubMed ID No1","Nuclear magnetic resonance spectroscopy (NMR) used ","Notes ","Ref_ID_No1"],axis=1)
```

```python
#Replaces n.a with its actuall definition and replace genus ocurence divider from "," to "/"


cyanodb=cyanodb.replace("n.a.",np.nan)                     
cyanodb["Genus"]=cyanodb["Genus"].str.replace(";","/")     
cyanodb
```
