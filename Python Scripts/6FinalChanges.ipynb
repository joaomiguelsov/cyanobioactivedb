```python
#POSSIVELMENTE PARA RETIRAR O QUE ESTA ABAIXO MENOS ULTIMA LINHA
```

```python
import unicodedata,ftfy

def fix_names(name):
    name_fix1 =ftfy.fix_text(name)
    name_fix2=unicodedata.normalize("NFKD", name_fix1)
    new_name="".join([letter for letter in name_fix2 if not unicodedata.combining(letter)])
    return new_name


new_names_list=[]
```

```python
for x in cyano["Compound_name"]:
    a= fix_names(str(x))
    print(a)
    new_names_list.append(a)
```

```python
cyano["Compound_name"]=new_names_list
```
