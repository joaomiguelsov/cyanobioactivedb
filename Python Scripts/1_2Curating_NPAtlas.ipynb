```python
#Splits the list by "," in order to select each individual genus.

genus_list=str(genus_list).title()
genus_list=genus_list.split(",")
genus_list
```

```python
#This command makes a new proper list of the cyanobacterial genus
#The first and last elements have to have the first and last symbols removed to avoid mistakes in this case.


genus_list2=[]
for genus in genus_list:
    if genus==genus_list[0]:
        genus_list2.append(genus[2:])
    elif genus==genus_list[-1]:
        genus_list2.append(genus[:-2])
    else:
        genus_list2.append(genus)
genus_list2
```

```python
atlas3 = atlas2[atlas2['genus'].isin(genus_list2)]      #Makes a new database that contains only the compounds from cyanobacteria
atlas3 = atlas3.reset_index(drop=True)                  #Resets the index again to avoid confusion
atlas3
```

```python
atlas3.to_csv("Atlas_final.csv",sep=";",index=False)      #saves the database to a csv file.
```
